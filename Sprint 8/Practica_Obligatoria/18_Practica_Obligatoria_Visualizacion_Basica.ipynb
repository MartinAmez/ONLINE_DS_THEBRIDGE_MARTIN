### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# HISTOGRAMA
fig2=plt.figure()
ejex=data.area_land_km2
font_dict={"family":"Gabriola",
           "size":16}
plt.hist(ejex, bins=60, alpha=0.5, label="Hist Area")
plt.xlabel(r"Superficie (km$^2$)",fontdict=font_dict) #me falta la negrita
plt.ylabel("Numero de ciudades",fontdict=font_dict)
plt.legend();
fig2.savefig("./img/imagen2.png")
```

```python
# COUNT PLOTS
fig1=plt.figure()
ejey=data.population_total.sort_values(ascending=False)[0:5]/1000
ejex=[
data.city[239],
data.city[367],
data.city[359],
data.city[9],
data.city[386]]

plt.bar(ejex,ejey,color=["blue","grey","brown","olive","pink"], alpha=0.5)
for i, valor in enumerate(ejey):
    plt.text(i, valor + 0.1, str(valor), ha='center', va='bottom')
plt.title("Ciudades más pobladas de California")
plt.ylim(0,5000)
plt.ylabel("Habitantes (miles)")
plt.xlabel("Ciudades");
fig1.savefig("./img/imagen1.png")
```

```python
# SCATTER PLOT
fig3=plt.figure()
plt.scatter(data.area_water_km2,data.population_total/1000, c="g", alpha=0.3)
plt.xlabel(r"Area de agua en la ciudad ($km^2$)")
x=data.area_water_km2
y=data.population_total/1000
o1_x=data.loc[data["population_total"].idxmax(), "area_water_km2"]
o1_y=y.max()
o2_x=x.max()
o2_y= data.loc[data["area_water_km2"].idxmax(), "population_total"]/1000
x_text=(60)
y_text=(3500)
font_dict={
    "size":16,
    "family":"Calibri",
    "color":"black"
}
props_flecha={
    "arrowstyle":"->",
    "connectionstyle":"arc"
}
plt.title("Población por kilómetro cuadrado de agua")
plt.annotate("Outliers", xy=(o1_x,o1_y), xytext=(x_text,y_text),arrowprops=props_flecha)
plt.annotate("Outliers", xy=(o2_x,o2_y), xytext=(x_text,y_text),arrowprops=props_flecha);
fig3.savefig("./img/imagen3.png")
```

```python
# BOXPLOT
fig4=plt.figure()
box_props=dict(facecolor="red")
median_props={"color":"black"}
plt.boxplot(data.elevation_m, vert=False, patch_artist=True, boxprops=box_props, medianprops=median_props)
plt.xlabel("Altura (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California");
fig4.savefig("./img/imagen4.png")
```
