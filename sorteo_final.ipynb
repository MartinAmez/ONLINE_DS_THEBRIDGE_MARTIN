```python
import pandas as pd
import random as rm

from collections import Counter
from pathlib import Path
```

```python
def print_in_columns(data, num_columns=5):
    for i, item in enumerate(data):
        # Print item followed by a tab, without a newline
        print(f"{item}\t", end='')
        # Every num_columns items, print a newline
        if (i + 1) % num_columns == 0:
            print()
    # Print a final newline in case the last line is not complete
    if len(data) % num_columns != 0:
        print()
```
